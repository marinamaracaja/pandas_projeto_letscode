## 1 - Carregar os arquivos e cadastro, investimento1 e investimento2 e mostrar a quantidade de linhas de cada um, além das informações desses dataframes

```python
#Importação de bibliotecas e de arquivos para conversão em dataframes
import pandas as pd
import datetime as dt
import numpy as np

cadastro = pd.read_csv('cadastro.csv')
inv1 = pd.read_csv('investimento_parte1.csv')
inv2 = pd.read_csv('investimento_parte2.csv')
```

## 2 - Juntar os dataframes de investimento1 e investimento2 em um só e ver a quantidade de linhas

```python
#Concatenação dos dataframes de investimentos e avaliação nas dimensões 
#do concatenado e do cadastro
inv = pd.concat([inv1, inv2])
inv.shape
```

```python
cadastro.shape
```

## 3 - Crie um novo dataframe eliminando as linhas duplicadas

```python
#Drop das duplicatas e verificação que agora o dataframe de cadastro e o de investimentos possuem o mesmo número de linhas
inv = inv.drop_duplicates()
inv.shape
```

```python
#Notou-se que o dataframe de cadastro já encontrava-se organizado por id de forma crescente.
cadastro.head()
```

## 4 - Junte o dataframe de investimento com o de cadastro
Dica: use a chave id

```python
#Fusão de ambos os dataframes, cadastro e investimento em um único dataframe.
DF = cadastro.merge(inv, on='id')
DF.head(10)
```

## 5 - Ordene o dataframe pelo id, de forma crescente e redefina o índice

```python
#Reorganização das linhas do dataframe de investimentos por id, resetando index
DF = DF.sort_values(by='id', ascending=True)
DF = DF.reset_index(drop=True)
DF.head()
```

## 6 - Faça uma análise dos dados faltantes e trate-os

Mostre todas as colunas com dados faltantes, sua quantidade/porcentagem e como tratar (preencher) cada uma delas explicando sua estratégia

```python
#Avaliar quantos são os faltantes no dataframe.
DF.isna().sum() # Como são apenas nos numéricos (Ativos e Notas), pode-se substituir por zero
```

```python
# Pode-se substituir os vazios por zero.
DF = DF.fillna(0)
DF.isna().sum()
```

## 7 - Manipule os dados a seguir:
- Criar coluna "Total" somando o valor de todos os ativos por investidor
- Criar coluna "Media" com a média das notas 1 e 2
- Criar coluna "Dominio" com apenas o domínio do email (dica: separar por @)
- Mudar o nome da coluna "Salário" para "Renda" e tratar a coluna (sem caracter $ e tipo numérica)
- Mudar os dados da coluna "Sexo" para Masculino e Feminino (tente usar o map)
- Transformar os dados da coluna Media em inteiro
- Criar coluna "NPS" onde o investidor é promotor se a média das notas é maior igual a 80, neutro se a média estiver entre 50 e 80, e detrator se for abaixo de 50


```python
#Criação da Coluna Total, somatório dos Ativos.
DF['Total'] = DF[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06', 'Ativo07']].sum(axis=1)
DF.head()
```

```python
#Criação da Coluna Média, média das notas.
DF['Media'] = DF[['Nota1', 'Nota2']].mean(axis=1)
DF.head()
```

```python
#Uso de função para separa a string de email em três partes e recuperar o domínio para adicionar numa nova coluna no Dataframe.
def dominio(i):
  i = i.split('@')
  i = i[1]
  i = i.split('.')
  return i[0]
DF['Dominio'] = DF['Email'].apply(dominio)
DF.head()
```

```python
#Uso de uma função criada para retirar do string o $ de Salario, e após isso, alterar o nome da coluna para Renda e os valores para float.
def renda(i):
  i = i.split('$')
  i = i[1]
  return i
DF['Salario'] = DF['Salario'].apply(renda)
DF = DF.rename(columns={'Salario':'Renda'})
DF['Renda'] = DF['Renda'].astype(float)
DF.head(3)
```

```python
#Uso da funcionalidade Map para modificar na coluna Sexo os termos, masc para masculino e fem para feminino.
DF['Sexo'] = DF['Sexo'].map({'Fem':'Feminino','Masc':'Masculino'})
DF.head()
```

```python
#Verificação de como a coluna Média se encontra no dataframe.
DF.dtypes
```

```python
#Por não estar como inteiro, e sim como float, o que gera muitas casas decimais e além de poluir, é desnecessário, converteu-se a coluna para inteiro.
DF['Media'] = DF['Media'].astype(int)
DF.dtypes
```

```python
#Criação de uma condicional com máscara em Numpy com uma ferramenta chamada select para criar a coluna NPS a partir do valor da Média dada.
condicao = [
    (DF['Media'] >= 80) ,
    (DF['Media'] >= 50) & (DF['Media'] <80),
    (DF['Media'] < 50)]
investidor = ['Promotor', 'Neutro', 'Detrator']
DF['NPS'] = np.select(condicao, investidor, default='Não especificado')
DF.head()
```

## 8 - Calcule os seguintes dados
- Calcule o total investido em cada coluna Ativo (tente usar o apply)
- Qual o menor valor investido em cada ativo?
- Mostre a média e desvio padrão de cada ativo
- Qual a idade do investidor mais velho?
- Qual a mediana da coluna Média?

```python
#Somatório de quanto há investido em cada ativo, observa-se que os menos investidos são o Ativo 7, 1 e 6 respectivamente.
Somatorios_Ativos = DF[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06', 'Ativo07']].sum(axis =0)
Somatorios_Ativos
```

```python
#O mínimo investido por algum investidor em cada ativo é demonstrado a seguir, nota-se que o único Ativo em que todos investem é o 7, apesar de ser um dos menos investidos, o que demonstra, provavelmente, uma rentabilidade baixa, porém garantida.
Ativos_minimos = DF[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06', 'Ativo07']].min(axis =0)
print(Ativos_minimos)
```

```python
#Demonstração dos valores obtidos da média e do desvio padrão dos investimentos de cada ativo, nota-se uma média extremamente alta em alguns valores, e extremamente baixa em outras (o que demonstra o quanto se rentabiliza em cada ativo, normalmente), já o desvio padrão demonstra o quanto varia o valor dos investimentos, sendo desvios altos demosntração de um alto rsico ou de uma confianção baixa no ativo específico.
Ativos_media = DF[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06', 'Ativo07']].mean(axis =0)
Ativos_desvio = DF[['Ativo01', 'Ativo02', 'Ativo03', 'Ativo04', 'Ativo05', 'Ativo06', 'Ativo07']].std(axis =0)
print(f'{Ativos_media}\n {Ativos_desvio}')
```

```python
#Percebe-se então que está se lidando com uma faixa de população economicamente ativa que ainda não parou de trabalhar e ainda está produzindo renda antes da aposentadoria, uma pesquisa de finalidade de investimentos poderia ser arranjada para busca de públicos alvo.
Mais_velho = DF['Idade'].max()
print(Mais_velho)
```

```python
#Tem-se uma mediana extremamente baixa para a média de notas, faz-se necessário avaliar o que há de errado com a plataforma ou se o sistema de notas precisa ser simplificado para um sistema de estrelas, por exemplo (inteiro já de início).
Mediana_Media = DF['Media'].median()
print(Mediana_Media)
```
